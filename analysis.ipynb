```python
def total_fuels(df):
    """
    This function creates two new columns, one for the total haleu and one for the total leup fuel.

    Parameters
    ----------
    df : pandas.DataFrame
        DataFrame containing the columns 'haleu' and 'leup'.

    Returns
    -------
    df : pandas.DataFrame
        DataFrame with two new columns: 'haleu_total' and 'leup_total'.
    """

    df['haleu_total'] = df['Total Xe100h'] + df['Total MMRh']
    df['leup_total'] = df['Total Xe100l'] + df['Total MMRl']
    df['uox_total'] = df['Total UOx'] + df['Total AP UOx']

    return df
```

```python
mb_g_fresh = pd.read_csv(mb+'greedy/fresh/dg2_fresh_fuel.csv')
mb_r_fresh = pd.read_csv(mb+'random/fresh/dr2_fresh_fuel.csv')
mb_rg_fresh = pd.read_csv(ob+'rand_greed/fresh/drg2_fresh_fuel.csv')

ob_g_fresh = pd.read_csv(ob+'greedy/fresh/dg2_fresh_fuel.csv')
ob_r_fresh = pd.read_csv(ob+'random/fresh/dr2_fresh_fuel.csv')
ob_rg_fresh = pd.read_csv(ob+'rand_greed_one/fresh/drg2_fresh_fuel.csv')
```

```python
mb_g_fresh  = total_fuels(mb_g_fresh)
mb_r_fresh  = total_fuels(mb_r_fresh)
mb_rg_fresh = total_fuels(mb_rg_fresh)

ob_g_fresh  = total_fuels(ob_g_fresh)
ob_r_fresh  = total_fuels(ob_r_fresh)
ob_rg_fresh = total_fuels(ob_rg_fresh)
```

```python
algorithm = ("Greedy", "Random", "Random Greedy")
fuel = {
    'HALEU': [round(mb_fresh['mb_g_haleu_total'][1751]/1e6,3),
              round(mb_fresh['mb_r_haleu_total'][1751]/1e6,3),
              round(mb_fresh['mb_rg_haleu_total'][1751]/1e6,3)],
    'LEU': [round(mb_fresh['mb_g_uox_total'][1751]/1e6,3),
            round(mb_fresh['mb_r_uox_total'][1751]/1e6,3),
            round(mb_fresh['mb_rg_uox_total'][1751]/1e6,3)],
    'LEUP': [round(mb_fresh['mb_g_leup_total'][1751]/1e6,3),
             round(mb_fresh['mb_r_leup_total'][1751]/1e6,3),
             round(mb_fresh['mb_rg_leup_total'][1751]/1e6,3)]
}

x = np.arange(len(algorithm))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in fuel.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Total Mass Fresh Fuel [kt]')
ax.set_xticks(x + width, algorithm)
ax.legend(loc='upper left', ncols=3, bbox_to_anchor=(0, 1.1))

ax.set_yscale('log')

plt.savefig('images/final_cumulative_fuel_mf')
```
